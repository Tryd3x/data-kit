```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
# load dataset
tips = sns.load_dataset("tips")
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
tips.info()
```

#  **Seaborn Fundamentals**

## scatterplot


```python
sns.scatterplot(x='tip', y = 'total_bill', data=tips,
                hue = 'day',
                palette = 'YlGnBu')
```

## lineplot

```python
sns.lineplot(x='tip', y = 'total_bill', data=tips,
                hue = 'sex')
```

## histplot


```python
sns.histplot(x = 'tip', data = tips,
             kde = False, bins = 15,
             )
```

## barplot


```python
sns.barplot(x = 'day', y = 'tip', data = tips,
            palette = 'YlGnBu', hue='sex',
            errorbar = None)
```

## boxplot


```python
sns.boxplot(x='day', y = 'tip', data=tips)
```

## countplot

```python
sns.countplot(data = tips, x='sex', hue='day')
```

## stripplot


```python
sns.stripplot(x='day', y='tip', data = tips,
              hue = 'sex',
              dodge = True)
```

## jointplot

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = tips,
              # kind = 'reg',
              kind = 'hex')
```

## heatmap

```python
titanic = sns.load_dataset("titanic")
num_col = titanic.select_dtypes(include=['int64','float64']).columns
```

```python
sns.heatmap(titanic[num_col].corr(), annot=True, cmap="coolwarm")
```

## clustermap

```python
iris = sns.load_dataset('iris')
sns.clustermap(iris.drop('species',axis=1))
```

## pointplot

```python
sns.pointplot(data = tips, x = 'day', y='tip')
```

## violinplot

```python
sns.violinplot(data=tips,x = 'day', y='tip')
```
